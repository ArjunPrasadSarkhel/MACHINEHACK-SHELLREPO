```python
import pandas as pd
```

```python
data = pd.read_json("/content/drive/MyDrive/MachineHACK_AI_COMP/datafinal.json")
data.head()
```

```python
def word_count(text):
    words = text.split()
    return len(words)

# Apply the function to each row in the "Text" column and store the result in a new column
data['WordCount'] = data['Text'].apply(word_count)
```

```python
data_F = data[(data['ContainsCode']==False)]
data_F.head()
```

```python
data_F
```

```python
# Reset the index if needed
data_F.reset_index(drop=True, inplace=True)
data_F.head()
```

```python
import re
# Regex pattern to detect code
code_pattern = r'\{[^\{\}]*\}|print|git|;|:|bash'

# Function to check if the code pattern exists in the given text
def has_code(text):
    return bool(re.search(code_pattern, text))

# Apply the 'has_code' function to the 'text_column' and create a new column 'has_code'
data_F['has_code'] = data_F['Text'].apply(has_code)
```

```python
data_F.drop(['has_code'], axis=1, inplace=True)
```

```python
data_all = pd.concat([data_T, data_F], ignore_index=True)
data_all.head()
```

```python
data_all = data_all.sample(frac=1, random_state=42)
```

```python
data_all.reset_index(drop=True, inplace=True)
```

```python
# Using Q Lora, 4bit
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
MODEL_NAME = "meta-llama/Llama-2-7b-hf"

model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    device_map="auto",
    trust_remote_code=True,
    quantization_config=bnb_config
)

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
tokenizer.pad_token = tokenizer.eos_token
```

```python
peft_config = LoraConfig(
    lora_alpha=32,
    lora_dropout=0.05,
    r=16,
    bias="none",
    task_type="CAUSAL_LM",
    target_modules=[
    "q_proj",
    "v_proj"]
)
```

```python
model = get_peft_model(model, peft_config)
```

```python
training_args = transformers.TrainingArguments(
      per_device_train_batch_size=1,
      gradient_accumulation_steps=4,
      num_train_epochs=2,
      learning_rate=2e-4,
      fp16=True,
      save_total_limit=3,
      logging_steps=1,
      output_dir="/content/drive/MyDrive/MachineHACK_AI_COMP/New_Trial_Above91/experiments",
      optim="paged_adamw_8bit",
      lr_scheduler_type="cosine",
      warmup_ratio=0.05,

)

trainer = SFTTrainer(
    model=model,
    train_dataset = data['train'],
    dataset_text_field = 'text',
    args=training_args,
    data_collator=transformers.DataCollatorForLanguageModeling(tokenizer, mlm=False),
    max_seq_length = 512
)
```

```python
from transformers import StoppingCriteria
class MyStoppingCriteria(StoppingCriteria):
  def __init__(self, target_sequence, prompt):
    self.target_sequence = target_sequence
    self.prompt=prompt

  def __call__(self, input_ids, scores, **kwargs):
      # Get the generated text as a string
      generated_text = tokenizer.decode(input_ids[0])
      generated_text = generated_text.replace(self.prompt,'')
      # Check if the target sequence appears in the generated text
      if self.target_sequence in generated_text:
          return True  # Stop generation

      return False  # Continue generation

  def __len__(self):
      return 1

  def __iter__(self):
      yield self
```
